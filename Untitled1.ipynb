```python
# LSTM for international airline passengers problem with regression framing
import numpy
import matplotlib.pyplot as plt
from pandas import read_excel
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error


numpy.random.seed(7)
# load the dataset
dataframe = read_excel('Data for LSTM.xlsx', index_col=0)
dataset = dataframe.values
dataset = dataset.astype('float32')

# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
dataset

train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

trainX = train[:,0:10].astype(float)
trainY = train[:,10]

testX = test[:,0:10].astype(float)
testY = test[:,10]


# reshape input to be [samples, time steps, features]
trainX = numpy.reshape(train, (train.shape[0], 1, train.shape[1]))
testX = numpy.reshape(test, (test.shape[0], 1, test.shape[1]))

# create and fit the LSTM network
def build_model():
    model = Sequential()
    model.add(LSTM(4, input_shape=(1, 11)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    return model

#K-Fold Cross Validation
k = 8
num_val_samples = len(trainX) // k
num_epochs = 200
all_scores = []
for i in range(k):
    print('processing fold #', i)
    val_data = trainX[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = trainY[i * num_val_samples: (i + 1) * num_val_samples]
    partial_train_data = numpy.concatenate([trainX[:i * num_val_samples],trainX[(i + 1) * num_val_samples:]],axis=0)
    partial_train_targets = numpy.concatenate([trainY[:i * num_val_samples],trainY[(i + 1) * num_val_samples:]],axis=0)
    
model = build_model()
model.fit(partial_train_data, partial_train_targets,
epochs=num_epochs, batch_size=1, verbose=1)
val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=1)
all_scores.append(val_mae)
print(all_scores)


# # evaluate the model
# scores = model.evaluate(trainX, trainY, verbose=0)
# print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# model.save("model.h5")
# print("Saved model to disk")
```
